```python
from azureml.core import Workspace, Dataset

subscription_id = '542e8f89-6341-4511-a2cf-537a546a3d57'
resource_group = 'MLAI'
workspace_name = 'pikahublah'

workspace = Workspace(subscription_id, resource_group, workspace_name)

dataset = Dataset.get_by_name(workspace, name='breastcancerdataset')
df = dataset.to_pandas_dataframe()
```

```python
# Z-Score: i need mu (avg) and sigma (std)

df.describe().T 
# T takes transpose of matrix- rows become cols, and cols become rows
```

```python
stats = df.describe().T
mu = stats['mean']
sigma = stats['std']
```
